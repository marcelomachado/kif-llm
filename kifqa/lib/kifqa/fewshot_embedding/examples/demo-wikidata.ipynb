```python
from kifqa.fewshot_embedding.embedding_serializer import EmbeddingSerializer
from kifqa.fewshot_embedding.loaders import JsonlLoader
from kifqa.model.example import Example
# from .autonotebook import tqdm as notebook_tqdm
```

```python
from typing import Optional


def parser(entry) -> Optional[Example]:
    try:
        input = str(entry['question'])
        output_str = ''
        output = []
        for triple in entry['triples']:
            if len(triple) == 3:
                output.append(f"""
    {{
        "subject": "{triple[0]}",
        "property": "{triple[1]}",
        "object": "{triple[2]}"
    }}""")
            inner = ',\n'.join(output)
            output_str = f"""
[
{inner}
]"""
        return Example(input, output_str)
    except Exception as e:
        print(f'could not parse entru {entry}, exception={e}')

output_path='/Users/marcelomachado/Documents/projects/kifqa/data-curated/annotated_wd_data_train_answerable_simple_max_1000_answers.pkl'
```

```python
serializer = EmbeddingSerializer(loader=JsonlLoader())
serializer.save_to_pickle(
    input_path='/Users/marcelomachado/Documents/projects/kifqa/data-curated/annotated_wd_data_train_answerable_simple_max_1000_answers.jsonl',
    output_path=output_path,
    parser_fn=parser)
```
